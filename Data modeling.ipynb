```python
import psycopg2
import pandas as pd
```

```python
def create_database():
    # connect to default database
    conn = psycopg2.connect("host=127.0.0.1 user=postgres dbname=postgres password=sonu")
    cur = conn.cursor()
    conn.set_session(autocommit=True)
    
    # create a database with UTF-8 encoding
    cur.execute("DROP DATABASE flights")
    cur.execute("CREATE DATABASE flights")
    
    # close the connection to the default databse
    conn.close()
    
    # connect to your new database
    conn = psycopg2.connect("host=127.0.0.1 user=postgres dbname=flights password=sonu")
    cur = conn.cursor()
    
    return conn, cur
```

```python
airline_review = pd.read_csv("AirlineScrappedReview.csv")
```

```python
airline_review.head()
```

```python
airline_review_clean = airline_review[['Passanger_Name', 'Route', 'Rating', 'Review_title', 'Class']]
```

```python
airline_review_clean.head()
```

```python
airline_review_clean.tail(20)
```

```python
airline_review_clean = airline_review_clean.dropna()
```

```python
airline_review_clean.tail(20)
```

```python
airline_review_clean =  airline_review_clean[~airline_review_clean['Route'].isin(['yes','Business Class','Economy Class','First Class','May 2015'])]
```

```python
airline_review_clean.head()
```

```python
conn, cur = create_database()
```

```python
flight_table_create = ("""CREATE TABLE IF NOT EXISTS flightsrating (
                      Passanger_Name varchar,
                      Route varchar,
                      Rating float,
                      Review_title varchar,
                      Class varchar)
                       """)
```

```python
cur.execute(flight_table_create)
```

```python
flight_table_insert = ("""INSERT INTO flightsrating(Passanger_Name, Route, Rating, Review_title, Class)
                        VALUES(%s, %s, %s, %s, %s)
                       """)
```

```python
for i, row in airline_review_clean.iterrows():
    cur.execute(flight_table_insert, list(row))
```

```python
conn.commit()
```

```python
cur.close()
conn.close()
```
